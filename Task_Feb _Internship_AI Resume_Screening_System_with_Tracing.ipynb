```python

# STEP 1: Job Description
job_description = """
Looking for Data Scientist with skills in Python, Machine Learning, NLP, SQL.
"""
# STEP 2: Resumes
resume_strong = """
Experienced Data Scientist with skills in Python, Machine Learning, NLP, SQL, Deep Learning.
"""

resume_average = """
Graduate with basic knowledge of Python and Machine Learning.
"""
resume_weak = """
Fresher with no experience in Data Science.
"""
# STEP 3: Skill Extraction
def extract_skills(text):
    skills_list = ["python", "machine learning", "nlp", "sql", "deep learning"]

    found = []
    for skill in skills_list:
        if skill in text.lower():
            found.append(skill)

    return found

# STEP 4: Matching + Scoring
def score_resume(resume, job):
    job_skills = extract_skills(job)
    resume_skills = extract_skills(resume)

    match_count = len(set(job_skills) & set(resume_skills))
    total = len(job_skills)

    score = int((match_count / total) * 100) if total > 0 else 0

    return score, resume_skills
# STEP 5: Explanation
def explain(score):
    if score >= 75:
        return "Strong candidate with most required skills."
    elif score >= 40:
        return "Average candidate with some relevant skills."
    else:
        return "Weak candidate with limited or no required skills."

# STEP 6: Full Pipeline
def process_resume(name, resume):
    print("\n==========================")
    print("Candidate:", name)

    skills = extract_skills(resume)
    print("Extracted Skills:", skills)

    score, _ = score_resume(resume, job_description)
    print("Score:", score)

    explanation = explain(score)
    print("Explanation:", explanation)


# STEP 7: Run All Candidates
process_resume("Strong Candidate", resume_strong)
process_resume("Average Candidate", resume_average)
process_resume("Weak Candidate", resume_weak)
```
